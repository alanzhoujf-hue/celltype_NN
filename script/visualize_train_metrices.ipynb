```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_stage1 = pd.read_csv("/home/users/z/zhouji/celltype_DNN/allen_data/train_metrics_stage1.csv")
df_stage2 = pd.read_csv("/home/users/z/zhouji/celltype_DNN/allen_data/train_metrics_stage2_pruned.csv")
```

```python
df_stage1.columns = df_stage1.columns.str.strip()
df_stage2.columns = df_stage2.columns.str.strip()
```

```python
fig, (ax1, ax2, ax3, ax4)=plt.subplots(4, 1, figsize=(14, 12))
#ax1
ax1.plot(df_stage1["epoch"], df_stage1["train_loss"], label="Train Loss", marker=".", markersize=4)
ax1.plot(df_stage1["epoch"], df_stage1["val_loss"], label="Val Loss", marker=".", markersize=4)
ax1.set_title("Loss over Epochs Stage 1")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss")
ax1.legend()
ax1.grid(True)
#ax2
ax2.plot(df_stage1['epoch'], df_stage1['train_acc'], label='Train Acc', marker='.', markersize=4)
ax2.plot(df_stage1['epoch'], df_stage1['val_acc'], label='Val Acc', marker='.', markersize=4)
ax2.set_title('Accuracy over Epochs Stage 1')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy')
ax2.legend()
ax2.grid(True)
#ax3
ax3.plot(df_stage2["epoch"], df_stage2["train_loss"], label="Train Loss", marker=".", markersize=4)
ax3.plot(df_stage2["epoch"], df_stage2["val_loss"], label="Val Loss", marker=".", markersize=4)
ax3.set_title("Loss over Epochs Stage 2")
ax3.set_xlabel("Epoch")
ax3.set_ylabel("Loss")
ax3.legend()
ax3.grid(True)
#ax4
ax4.plot(df_stage2['epoch'], df_stage2['train_acc'], label='Train Acc', marker='.', markersize=4)
ax4.plot(df_stage2['epoch'], df_stage2['val_acc'], label='Val Acc', marker='.', markersize=4)
ax4.set_title('Accuracy over Epochs Stage 2')
ax4.set_xlabel('Epoch')
ax4.set_ylabel('Accuracy')
ax4.legend()
ax4.grid(True)

plt.tight_layout()
plt.show() #after save
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(14, 12))
#ax1
ax1.plot(df_stage1["epoch"], df_stage1["train_loss"], label="Train Loss", marker=".", markersize=4)
ax1.plot(df_stage1["epoch"], df_stage1["val_loss"], label="Val Loss", marker=".", markersize=4)
ax1.set_title("Loss over Epochs Stage 1")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss")
ax1.legend()
ax1.grid(True)
#ax2
ax2.plot(df_stage1['epoch'], df_stage1['train_acc'], label='Train Acc', marker='.', markersize=4)
ax2.plot(df_stage1['epoch'], df_stage1['val_acc'], label='Val Acc', marker='.', markersize=4)
ax2.set_title('Accuracy over Epochs Stage 1')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy')
ax2.legend()
ax2.grid(True)
#ax3
ax3.plot(df_stage2["epoch"], df_stage2["train_loss"], label="Train Loss", marker=".", markersize=4)
ax3.plot(df_stage2["epoch"], df_stage2["val_loss"], label="Val Loss", marker=".", markersize=4)
ax3.set_title("Loss over Epochs Stage 2")
ax3.set_xlabel("Epoch")
ax3.set_ylabel("Loss")
ax3.legend()
ax3.grid(True)
#ax4
ax4.plot(df_stage2['epoch'], df_stage2['train_acc'], label='Train Acc', marker='.', markersize=4)
ax4.plot(df_stage2['epoch'], df_stage2['val_acc'], label='Val Acc', marker='.', markersize=4)
ax4.set_title('Accuracy over Epochs Stage 2')
ax4.set_xlabel('Epoch')
ax4.set_ylabel('Accuracy')
ax4.legend()
ax4.grid(True)

plt.tight_layout()
plt.savefig('/home/users/z/zhouji/celltype_DNN/allen_data/metrics_plot_stage_1_2.png')
plt.show() #after save
```

```python
best_val_acc_stage1 = df_stage1['val_acc'].max()
best_val_acc_epoch_stage1 = df_stage1.loc[df_stage1['val_acc'].idxmax(), 'epoch']

min_val_loss_stage1 = df_stage1['val_loss'].min()
min_val_loss_epoch_stage1 = df_stage1.loc[df_stage1['val_loss'].idxmin(), 'epoch']

final_train_acc_stage1 = df_stage1['train_acc'].iloc[-1]
final_val_acc_stage1 = df_stage1['val_acc'].iloc[-1]

best_val_acc_stage2 = df_stage2['val_acc'].max()
best_val_acc_epoch_stage2 = df_stage2.loc[df_stage2['val_acc'].idxmax(), 'epoch']

min_val_loss_stage2 = df_stage2['val_loss'].min()
min_val_loss_epoch_stage2 = df_stage2.loc[df_stage2['val_loss'].idxmin(), 'epoch']

final_train_acc_stage2 = df_stage2['train_acc'].iloc[-1]
final_val_acc_stage2 = df_stage2['val_acc'].iloc[-1]
```

```python
print(f"Total Epochs Stage1: {len(df_stage1)}")
print(f"Best Validation Accuracy Stage1: {best_val_acc_stage1:.4f} at Epoch {best_val_acc_epoch_stage1}")
print(f"Lowest Validation Loss Stage1: {min_val_loss_stage1:.4f} at Epoch {min_val_loss_epoch_stage1}")
print(f"Final Train Accuracy Stage1: {final_train_acc_stage1:.4f}")
print(f"Final Validation Accuracy Stage1: {final_val_acc_stage1:.4f}")
print(df_stage1.tail().to_string(index=False))

print(f"Total Epochs Stage2: {len(df_stage2)}")
print(f"Best Validation Accuracy Stage2: {best_val_acc_stage2:.4f} at Epoch {best_val_acc_epoch_stage2}")
print(f"Lowest Validation Loss Stage2: {min_val_loss_stage2:.4f} at Epoch {min_val_loss_epoch_stage2}")
print(f"Final Train Accuracy Stage2: {final_train_acc_stage2:.4f}")
print(f"Final Validation Accuracy Stage2: {final_val_acc_stage2:.4f}")
print(df_stage2.tail().to_string(index=False))
```
